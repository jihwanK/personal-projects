```python
import requests
import time
import pandas as pd
```

```python
url = "http://api.springernature.com/meta/v2/json"
headers = {
    'Accept': 'application/json',
}
log = f''
start = 1
records = []
result = {
    'title': [], 
    'abstract': [],
    'subjects': [],
}
```

```python
for idx in range(100):
    params = {
        'q': 'type:Journal',
        'p': 100,
        's': start,
        'api_key': '793eb0a9919da0a054f623d6fec08c4a'
    }

    response = requests.get(url, headers=headers, params=params)
    data = response.json()

    data_record = data['records']
    for record in data_record:
        result['title'].append(record['title'])
        result['abstract'].append(record['abstract'])
        result['subjects'].append(','.join(record['subjects']))
    records.extend(data_record)

    start += 100
    
    print(f'[{idx+1}] {start} articles have been crawled')

    time.sleep(1)
```

```python
df = pd.DataFrame(columns=['title', 'abstract', 'subjects'], data=result)
df
```
